## Plotly Example

The Plotly Python graph library makes interactive and professional quality visualizations. It is a popular tool for data exploration and can be easily be viewed in Notebooks.

Below is an example of sample code that will illustrate Plotly support through Notebooks.

```python
import plotly.graph_objs as go 
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot 
import numpy as np 
init_notebook_mode(connected=True) 

x = np.random.randn(2000) 
y = np.random.randn(2000) 
iplot([go.Histogram2dContour(x=x, y=y, contours=dict(coloring='heatmap')), 
go.Scatter(x=x, y=y, mode='markers', marker=dict(color='white', size=3, opacity=0.3))], show_link=False) 
```
